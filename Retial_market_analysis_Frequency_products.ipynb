## Performing set operations on the dataframe (A-B) , (A intersection B) for the first column items with  other items brought

```python
items = (df[0].unique())

encoded_vals = []
for index, row in df.iterrows():
    labels = {}
    uncommons = list(set(items) - set(row))
    commons = list(set(items).intersection(row))
    for uc in uncommons:
        labels[uc] = 0
    for com in commons:
        labels[com] = 1
    encoded_vals.append(labels)
encoded_vals[0]
```

## Converting in to a dataframe

```python
ohe=pd.DataFrame(encoded_vals)
```

```python
ohe.head()
```

## Frequent item sets 
## Apriori Algorithm application with support=0.01%

```python
freq_items= apriori(ohe,\
                     min_support=0.01,use_colnames=True,\
                     verbose=2)
freq_items.head()
```

## Rules are based on Confidence with min_threshhold=0.5

```python
rules = association_rules(freq_items, 
                          metric="confidence", 
                          min_threshold=0.5)
```

## Even though there is less support for items.
## High confidence and lift is required to check the item combinations.

```python
rules[rules.lift>=2]
```
